```python
import numpy as np
```

Return the value of Cosine

```python
def cosine():
    return(np.cos(x))
```

Return the value of Sine

```python
def sine():
    return(np.sin(x))
```

```python
A = 2*np.pi
B = A/1000
for x in np.arange(0,A,B):
    print("x = ",x,". sin(x) = ", sine(),". cos(x) = ",cosine(),".")
```
